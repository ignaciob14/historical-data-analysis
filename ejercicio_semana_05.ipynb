```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de los ingresos de Tesla (ajusta al sitio específico de los datos)
url = "https://es.wikipedia.org/wiki/Tesla,_Inc."
# Obtener el contenido de la página
html_data = requests.get(url).text

# Analizar el contenido HTML
soup = BeautifulSoup(html_data, 'html.parser')

# Extraer la tabla de ingresos (ajusta según la estructura de la página)
# Aquí se muestra cómo obtener la primera tabla, pero puedes ajustar al selector correcto
income_table = soup.find("table")  # Ajusta este selector según la página

# Convertir la tabla HTML en un DataFrame de Pandas
tesla_revenue = pd.read_html(str(income_table))[0]

# Mostrar las últimas cinco filas del DataFrame
print(tesla_revenue.tail())
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Crear el objeto Ticker para Tesla
tesla = yf.Ticker("TSLA")

# Descargar datos históricos de Tesla
tesla_data = tesla.history(period="max")

# Restablecer el índice del DataFrame
tesla_data.reset_index(inplace=True)

# Definir la función make_graph para graficar los datos
def make_graph(data, title="Tesla Stock Price Over Time"):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Close Price')
    plt.xlabel('Date')
    plt.ylabel('Close Price (USD)')
    plt.title(title)
    plt.legend()
    plt.show()

# Llamar a la función make_graph con los datos de Tesla
make_graph(tesla_data)
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Crear el objeto Ticker para GameStop
gamestop = yf.Ticker("GME")

# Descargar datos históricos de GameStop
gme_data = gamestop.history(period="max")

# Restablecer el índice del DataFrame
gme_data.reset_index(inplace=True)

# Definir la función make_graph para graficar los datos
def make_graph(data, title="Stock Price Over Time"):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Close Price')
    plt.xlabel('Date')
    plt.ylabel('Close Price (USD)')
    plt.title(title)
    plt.legend()
    plt.show()

# Llamar a la función make_graph con los datos de GameStop
make_graph(gme_data, title="GameStop Stock Price Over Time")
```
